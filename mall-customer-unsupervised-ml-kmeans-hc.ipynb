## **Calling Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.preprocessing as prs
import seaborn as sns
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
```

# **Show Data**

```python
df=pd.read_csv('/kaggle/input/marketing-data-for-a-supermarket-in-united-states/supermarket_marketing/Supermarket_CustomerMembers.csv')
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

## **Drop Unused Columns**

```python
df=df.drop(columns='CustomerID',axis=1)
```

```python
df2=df.copy()
```

## **Encoding**

```python
df2=df2.apply(LabelEncoder().fit_transform)
```

```python
df2=df2.to_numpy()
```

```python
x1 = df2[:, 0]
x2 = df2[:, 1]
x3 = df2[:, 2]
x4 = df2[:, 3]
x = np.vstack((x1, x2, x3, x4)).T
x
```

# **Modeling**
- ## Kmeans
- ## HC

# **KMeans**

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x)
    print('Cost_Function=',kmeans.inertia_,'with', i, 'Clusters')
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model = KMeans(n_clusters=5)
model.fit(x)
```

```python
clbl = model.labels_
clbl
```

```python
df['Cluster']=clbl
```

```python
df
```

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(x)
x_pca
```

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=clbl, cmap='Pastel2_r')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# **HC**

```python
import scipy.cluster.hierarchy as HC
from sklearn.cluster import AgglomerativeClustering
```

## **the methods** 
- single
- complete
- average
- ward
- centroid

## **Single**

```python
plt.figure(figsize=(15, 10))
HC.dendrogram(HC.linkage(x, method='single'))
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=6, linkage='single')
lbl = model.fit_predict(x)
```

```python
lbl
```

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=lbl, cmap='vlag')
plt.title("HCA Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

I think it's not good.

```python
plt.figure(figsize=(15, 10))
HC.dendrogram(HC.linkage(x, method='complete'))
plt.show()
```
